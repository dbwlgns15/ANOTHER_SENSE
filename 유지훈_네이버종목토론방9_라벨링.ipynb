```python
import pandas as pd
import numpy as np
from collections import Counter

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./src/네이버종토방댓글_카카오_15000_태그토큰화.csv')
```

```python
df['공포탐욕'] = 0.5
df
```

```python
def make_feargreed(df,fear,greed):
    label_list = df['공포탐욕'].to_list()
    token_list = df['토큰화댓글'].to_list()
    for i in range(len(token_list)):
        x = eval(token_list[i])
        if label_list[i] in [0,1]:
            continue
        if greed in x:
            label_list[i] += 0.5
            label_list[i] = int(label_list[i])
        if fear in x:
            label_list[i] -= 0.5
            label_list[i] = int(label_list[i])
    return label_list
```

```python
feargreed_wordset = [('하락','상승'),('손절','익절'),('악재','호재'),('폭락','폭등'),('떡락','떡상'),('매도','매수'),
                     ('안티','찬티'),('공포','탐욕'),('음봉','양봉'),('하한','상한'),('떨어지다','오르다'),
                     ('털다','줍다'),('내려가다','올라가다'),('죽다','살다')]
```

```python
for i in feargreed_wordset:
    fear, greed = i
    df['공포탐욕'] = make_feargreed(df,fear,greed)
    print(f'{i}체크 완료')
```
