```python
import pandas as pd
import numpy as np
import requests
import re
from bs4 import BeautifulSoup

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./src/네이버금융종목뉴스링크_네이버_전체.csv')
url_list = df['링크'].to_list()
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.41 Safari/537.36 Edg/101.0.1210.32'}

art_list = []
for i in range(len(url_list)):
    url = url_list[i]
    res = requests.get(url, headers = headers)
    bs = BeautifulSoup(res.text, 'html.parser')
    art_list.append(bs.find_all('div',{'class':'scr01'})[0].text)
    
    if (i+1) % 100 == 0:
        print(f'현재 {i+1}개 크롤링 완료')
```

```python
print(df.shape[0])
print(len(art_list))
df['기사'] = art_list
df
```

```python
df.to_csv("./src/네이버금융종목뉴스기사본문_네이버_전체.csv", index=False)
```
