```python
import pandas as pd
```

```python
df = pd.read_csv('./src/유사기사_제거(유사도기준-0.6).csv')
df = df[['날짜','제목','정보제공','기사']]
df = df.set_index('제목')
df
```

```python
df2 = pd.read_csv('./src/네이버금융종목뉴스기사본문_카카오_번역후.csv')
df2 = df2[['제목','링크','영문번역']]
df2 = df2.drop_duplicates('제목')
df2 = df2.set_index('제목')
df2
```

```python
df3 = df.join(df2).reset_index()
df3
```

```python
df3.to_csv("./src/유사기사_제거(유사도기준-0.6)+번역.csv", index=False)
```
