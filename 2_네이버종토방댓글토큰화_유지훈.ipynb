```python
import pandas as pd
import numpy as np
import re

import warnings
warnings.filterwarnings('ignore')
```

```python
comment_df = pd.read_csv('./src/네이버종토방댓글.csv')
comment_df.head()
```

```python
co_list = comment_df['댓글'].to_list()
co_list[:5]
```

```python
len1_list = []
for i in co_list:
    if len(str(i)) <= 1:
        i = re.sub('[^ㄱ-ㅎ가-힣]','',i)
        len1_list.append(i)
    if len(str(i)) == 2:
        i = re.sub('[^ㄱ-ㅎ가-힣a-zA-Z]','',i)
        if len(str(i)) <= 1:
            i = re.sub('[^ㄱ-ㅎ가-힣]','',i)
            len1_list.append(i)
print(len(len1_list))
print(len(set(len1_list)))
print(set(len1_list))
```

```python
len2_list = []
for i in co_list:
    if len(str(i)) == 2:
        len2_list.append(i)
print(len(len2_list))
print(len(set(len2_list)))
print(set(len2_list))
```
