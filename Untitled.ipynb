```python
df = stock.get_market_ohlcv("20210601", "20220601", "035720", adjusted=False)
df2 = stock.get_market_trading_volume_by_date("20210601", "20220601", "035720")
df = df.join(df2)
df = df[['등락률','개인','기관합계']]
df
```

```python
a = []
b = []
c = []
for i in df['등락률'].to_list():
    if i > 0:
        a.append(1)
    elif i < 0:
        a.append(-1)
    else:
        a.append(-1)
for i in df['개인'].to_list():
    if i > 0:
        b.append(1)
    elif i < 0:
        b.append(-1)
    else:
        b.append(0)
for i in df['기관합계'].to_list():
    if i > 0:
        c.append(1)
    elif i < 0:
        c.append(-1)
    else:
        c.append(0)
df['등락률'] = a
df['개인'] = b
df['기관합계'] = c
df
```
